```python
vid=cv2.VideoCapture(0)
while True:
    suc,img=vid.read()
    print(img.shape)
    cv2.imshow("capture",img)
    if cv2.waitKey(1) & 255 ==ord("o"):
        break

cv2.destroyAllWindows()
```

```python
vid=cv2.VideoCapture(0)
while True:
    suc,img=vid.read()
    #print(img.shape)
    cv2.rectangle(img,(270,200),(500,450),(0,0,255),1)
    crop_img=img[200:270,450:500]
    cv2.imshow("capture",img)
    cv2.imshow("crop_capture",crop_img)
    if cv2.waitKey(1) & 255 ==ord("o"):
        break

cv2.destroyAllWindows()
```
