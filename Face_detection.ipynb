# Process each zip file in the folder
for file_name in os.listdir(zip_folder):
    file_path = os.path.join(zip_folder, file_name)  # Full path to the file
    
    # Check if it's a zip file using zipfile.is_zipfile
    if zipfile.is_zipfile(file_path):
        extract_to = os.path.join(output_folder, file_name)  # Sub-folder for each file
        
        os.makedirs(extract_to, exist_ok=True)  # Create the folder
        
        # Extract the zip file
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)

### split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,class_label,test_size=0.20,stratify=class_label)
```

```python
xf_train,x_cv,yf_train,y_cv=train_test_split(x_train,y_train,test_size=0.20,stratify=y_train)
```

### normalization

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
norm=StandardScaler()
n_xf_train=norm.fit_transform(xf_train)
n_xf_cv=norm.transform(x_cv)
n_xf_test=norm.transform(x_test)
```

### model selection

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model
```

### training the model

```python
svm_model.fit(n_xf_train, yf_train)
```

### prediction on cross validation data

```python
y_pred_cv = svm_model.predict(n_xf_cv)
```

```python
from sklearn.metrics import classification_report,precision_recall_fscore_support
```

```python
print(classification_report(y_cv,y_pred_cv))
```
